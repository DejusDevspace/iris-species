## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the Dataset

```python
dataset = pd.read_csv('dataset/IRIS.csv')
```

## Understanding the Dataset

```python
print('The dataset has {} rows and {} columns.'.format(dataset.shape[0], dataset.shape[1]))
dataset.head()
```

```python
dataset.tail()
```

```python
# Checking for null or NaN values in the dataset
print(dataset.isnull().values.any())
print(dataset.isna().values.any())
```

```python
# Checking the descriptive statistics of the dataset
dataset.describe()
```

```python
# Visualizing the distribution of the species in the dataset
plt.figure(figsize=(5, 3), dpi=200)
plt.title('Data Distribution')
plt.xlabel('Species')
plt.ylabel('Number of Samples')
sns.countplot(x=dataset['species'], hue=dataset['species'])
plt.show()
```

The dataset is balanced as all species have equal number of entries in the dataset

```python
dataset.sample(6)
```

```python
# Getting the features each species
virginica = dataset[dataset['species'] == 'Iris-virginica']
versicolor = dataset[dataset['species'] == 'Iris-versicolor']
setosa = dataset[dataset['species'] == 'Iris-setosa']
```

```python
# Understanding the descriptive statistics of the iris-virginica species
virginica.describe()
```

```python
# Understanding the descriptive statistics of the iris-versicolor species
versicolor.describe()
```

```python
# Understanding the descriptive statistics of the iris-setosa species
setosa.describe()
```

NO sign of outliers or unusual range of values in each of the species<br>The dataset can be passed as clean...
